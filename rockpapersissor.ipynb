# Game of Rock, Paper & Sissor
---

|Player 1    |   Player 2|  Result|
|------------|-----------|--------|
|Rock        |Sissor     |P1      |
|Paper       |Rock       |P1      |
|Sissor      |Paper      |P1      |

```python
materials = {1:"Rock", 2:"Paper", 3:"Sissor"}
```

```python
class player:

    def __init__(self, material):
        self.score = 0
        self.material = material
    
player1 = player(materials[int(input("Select from the following: \n [1] Rock \n [2] Paper \n [3] Sissor"))])
player2 = player(materials[int(input("Select from the following: \n [1] Rock \n [2] Paper \n [3] Sissor"))])
        
```

```python
def game():
    winnings = [('Paper', 'Rock'), ('Rock', 'Sissor'), ('Sissor', 'Paper')]    
    for winning in winnings:
        if player1.material != player2.material:
            if player1.material in winning and player2.material in winning:
                if player1.material == winning[0]:
                    player1.score += 1
                    print(f'Player 1 is the winner with {player1.material} against player2 having {player2.material}, \n Score: Player1 [ {player1.score} ] vs Player2 [ {player2.score} ]')
                else:
                    player2.score += 1
                    print(f'Player 2 is the winner with {player2.material} against player1 having {player1.material}, \n Score: Player1 [ {player1.score} ] vs Player2 [ {player2.score} ]')

        else:
            print(f'Draw \n Score: Player1 [ {player1.score} ] vs Player2 [ {player2.score} ]')
            break
```

```python
game()
```
